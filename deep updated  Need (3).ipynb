```python
def feature(df_train):
    df_train['vpd']=df_train['Temperature_C']*(1-(df_train['Humidity']/100))
    df_train['Moisture_Retention'] = df_train['Soil_Moisture'] * df_train['Organic_Carbon']
    df_train['Moisture_Deficit'] = df_train['Humidity'] - df_train['Soil_Moisture']
    df_train['Water_Stress'] = df_train['Temperature_C'] / (df_train['Soil_Moisture'] + 1)
    df_train['Evaporation_Index'] = df_train['Temperature_C'] * df_train['Wind_Speed_kmh']
    df_train["Moisture_to_Heat_Ratio"] = df_train["Soil_Moisture"] / (df_train["Temperature_C"] + 1)

    return df_train
```

```python
df_train = feature(df_train)
df_test = feature(df_test)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df_train.corr(numeric_only=True),annot=True, cmap='Blues')
plt.title("Correlation between Features")
plt.show()
```

```python
x = df_train.drop(['Irrigation_Need'], axis=1)
y = df_train['Irrigation_Need']
y = to_categorical(y)
```

```python
ordinal_encoder = OrdinalEncoder()
cat_cols=list(x.select_dtypes( include=['object']).columns)
encoder=OrdinalEncoder()
x[cat_cols]=encoder.fit_transform(x[cat_cols])
df_test[cat_cols]=encoder.transform(df_test[cat_cols])
```

```python
len(x.columns), len(df_test.columns)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x= sc.fit_transform(x)
df_test=sc.transform(df_test)
```

```python
model = Sequential([
   Dense(128, activation='relu', input_shape=(x.shape[1],)),
   Dense(64, activation='relu'),
   Dropout(0.2),
   Dense(32, activation='relu'),
   Dense(3, activation='softmax')
])

model.summary()
```

```python
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

early_stopping = EarlyStopping(monitor='val_loss', patience=5)

checkpoints = ModelCheckpoint("best_model.keras", 
                              monitor="val_loss", 
                              save_best_only=True)

model.fit(x, y, epochs=50,
          validation_split=0.1, batch_size=100, 
          callbacks=[early_stopping])
```

```python
model.evaluate(x, y)
```

```python
model.save('model.h5')
```

```python
from sklearn.metrics import balanced_accuracy_score
pred_train = model.predict(x)
balanced_accuracy_score(pred_train.argmax(axis=1), y)
```
